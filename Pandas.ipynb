```python
points = {
    '1st Point' : [(3, 9), (0, 1), (3, 7), (5, 5)],
    '2nd Point' : [(4, -1), (1, -2), (5, -7), (0, -2)]
}
points = pd.DataFrame(points)
points
```

```python
# apply function on df
def distance(points):
    point_A = points['1st Point']
    point_B = points['2nd Point']
    return round(((point_A[0]-point_B[0])**2 + (point_A[1]-point_B[1])**2)**0.5, 2)

points.apply(distance, axis=1)
```

## Pandas GroupBy

```python
imdb_movies = pd.read_csv('./Datasets/imdb-top-1000.csv')
ipl = pd.read_csv('./Datasets/ipl-matches.csv')
```

```python
imdb_movies.head()
```

```python
imdb_movies.info()
```

```python
# groupby is applied to categorical column
# based on genres
genres = imdb_movies.groupby('Genre')
genres
```

```python
genres['IMDB_Rating'].mean()
```

```python
# top 3 genres by total earning
genres['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
top3_revenue = imdb_movies.groupby('Genre')['Gross']
top3_revenue.sum().sort_values(ascending=False).head(3)
```

```python
# Find the director with most votes
imdb_movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
imdb_movies.groupby('Genre')[['Series_Title', 'IMDB_Rating']].max()
```

```python
len(genres) # no.of groups
```

```python
genres.size() # no. of rows on each group
```

```python
# genres.first() -> 1st row of each group
# genres.last() -> last row of each group
# genres.nth(x) -> nth row of each group(0-index based)
```

```python
# all rows of a group
genres.get_group('Family')
```

```python
genres.groups # index position of rows
```

```python
genres.describe()
```

```python
# genres.head() -> first n rows of each group
# genres.tail() -> last n rows of each group
# genres.sample() -> random row form each group
```

```python
genres.sum()
temp = imdb_movies.groupby('Genre')[['Runtime', 'IMDB_Rating', 'No_of_Votes','Metascore']]
temp.agg(['min', 'max', 'mean'])
```

```python
temp.agg({
    'IMDB_Rating' : 'mean',
    'No_of_Votes' : 'sum',
    'Metascore' : 'min',
    'Runtime' : ['max', 'min']
})
```

```python
# loops in groups
for genre, df in genres:
    print(type(genre), type(df), end=" ")
    print(genre)
```

```python
# find number of movies staring with letter A, using apply fnction
def starting_with_A(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(starting_with_A)
```

```python
# movies rank on each group
def rank_movies_genre(group):
    group['Rank_on_Genre'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movies_genre)
```

```python
# groupby on multiple columns
duo = imdb_movies.groupby(['Director','Star1'])
```

```python
duo.size().sort_values(ascending=False).head(10)
```

```python
duo.get_group(('Peter Jackson', 'Elijah Wood'))
```

```python
# most earning duo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

## Mergin-Joining-Concatenating

```python
courses = pd.read_csv('./Datasets/courses.csv')
students = pd.read_csv('./Datasets/students.csv')
nov = pd.read_csv('./Datasets/reg-month1.csv')
dec = pd.read_csv('./Datasets/reg-month2.csv')
matches = pd.read_csv('./Datasets/ipl-matches.csv')
delivaries = pd.read_csv('./Datasets/deliveries.csv')
```

```python
nov.head()
```

```python
dec.head()
```

```python
# pd.concatenate
multi = pd.concat([nov, dec], keys=['nov', 'dec']) # multi index df
```

```python
multi.loc[('nov', 8)]
```

```python
pd.concat([nov, dec], axis=1) # horizontally
```

```python
# merging
# inner join
reg = pd.concat([nov, dec], ignore_index=True)
students.merge(reg, how='inner', on='student_id') # only common
```

```python
# left join
courses.merge(reg, how='left', on='course_id')
```
