# Pandas

### Pandas Series
- 1D array holding data of any type. Pandas series is like a cloumn of a table.

```python
import numpy as np
import pandas as pd
```

```python
# using list
country = ['India', 'Nepal', 'Bhutan', 'USA', 'Russia']
countries = pd.Series(country)
countries
```

```python
marks = [97, 94, 82, 77]
subjects = ['Eng', 'Math', 'Ben', 'Phy']

# custom indices
pd.Series(marks, index=subjects, name='My marks')
```

```python
# using dictionaries
marks = {
    'Math' : 94,
    'Phy' : 77,
    'Chem' : 71,
    'Bio' : 82
}
my_marks = pd.Series(marks, name='My marks')
my_marks
```

```python
temp = pd.Series([1, 1, 2, 9, 9, 2, 2, 7, 4, 8])
temp.drop_duplicates()
```

```python
# by default it is taking first value, but it can be changed using keep parameter
temp.drop_duplicates(keep='last')
```

```python
temp.duplicated().sum()
```

```python
# is nun
temp = pd.Series([2, 1, np.nan, 0, np.nan, 0, np.nan, 88, 9])
temp
```

```python
temp.isnull().sum()
```

```python
temp.size # returns number of elements
```

```python
temp.count() # count values except null values
```

```python
# dropna -> removes all missing values
temp.dropna()
```

```python
# fillna -> fill missing value
temp.fillna(np.round(temp.mean(),2))
```
