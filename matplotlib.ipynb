```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2D Plot

```python
rs_vk_score = pd.read_csv('https://drive.google.com/uc?id=1vI6QgVkD4UZHlTEknKDHJGkdwdbmMx__')
rs_vk_score
```

```python
plt.plot(rs_vk_score['index'], rs_vk_score['V Kohli'], color='red', linestyle='dashed')
# linestyle -> solid, dashed, dotted, dashdot
plt.plot(rs_vk_score['index'], rs_vk_score['RG Sharma'], color='#946B63', linewidth=3) # hexcode

plt.title('VK-RS Total Score in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(rs_vk_score['index'], rs_vk_score['V Kohli'], marker='D', label='VK')
plt.plot(rs_vk_score['index'], rs_vk_score['RG Sharma'], marker='>', markersize=10, label='RS')

plt.title('VK-RS Total Score in IPL')
plt.xlabel('Season')
plt.ylabel('Run Scored')

# trim
plt.xlim(2010, 2014)
plt.ylim(300, 700)

plt.legend(loc='best')
# location -> 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

plt.grid()
plt.show()
```

## Scatter Plot

```python
batsman = pd.read_csv('https://drive.google.com/uc?id=1kJ1v2GDxa5uADs3uproeMp588RtRggJ_')
top_50_batsman = batsman.sort_values('runs', ascending=False).head(50)
```

```python
top_50_batsman
```

```python
plt.scatter(top_50_batsman['avg'], top_50_batsman['strike_rate'], marker='o', color='red', s=top_50_batsman['runs']/50)
# plt.plot(x, y, 'o') faster but limited
plt.title('Avg - StrikeRate')
plt.xlabel('AVG Runs')
plt.ylabel('Strike Rate')

plt.grid()
plt.show()
```
